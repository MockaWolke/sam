```python
from scipy.stats import beta, norm
import matplotlib.pyplot as plt
import numpy as np
```

```python

# p,q or ame
params = [3.2, (3.23,2.39)]
```

```python
plt.figure(figsize=(14, 5))
x = np.linspace(0, 1, 100)



def plot(func):

    for p in params:
        
        if isinstance(p, float):
        
            plt.plot(x, func(x, p, p), label = f"p = {p}, q = {p}")
        else:
            plt.plot(x, func(x, p[0], p[1]), label = f"p = {p[0]}, q = {p[1]}")

plt.subplot(131)
plot(beta.pdf)
plt.title("Dichte")
plt.subplot(132)
plot(beta.cdf)
plt.title("Verteilungs")

plt.subplot(133)

plot(beta.ppf)
plt.title("Quantilsfunktion")
plt.legend()
plt.show()
```
